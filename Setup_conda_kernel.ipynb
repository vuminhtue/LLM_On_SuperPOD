# Setup working env in SuperPOD
## 1. Setup Conda env
### 1.1 Request compute node
- Open Cursor's Terminal and login to SuperPOD's login node

```
$ ssh username@superpod.smu.edu
```

- Request for a compute node: If you already have your own ColdFront allocation feel free to use yours:

```
$ srun -A your_own_alloc -N1 -c10 -G1 --mem=16gb --time=2:00:00 --pty $SHELL
```

- If you do not have ColdFront allocation, I have added you to my ColdFront allocation. You need to accept EULA sent to your email:

```
$ srun -A tuev_oitrts_workshop_0002 -N1 -c10 -G1 --mem=16gb --time=2:00:00 --pty $SHELL
```
or
```
srun -A tuev_oitrts_workshop_0002 -N1 -c10 -G1 --mem=16gb --time=2:00:00 --reservation=oit_llm_workshop --pty $SHELL
```

**Note**: the flag *--reservation=oit_llm_workshop* is only available for this workshop to secure a node for you and will be cancel later.

### 1.2 Load module with GPU support and create conda env

```
$ module load conda gcc/11 cuda/11 cudnn/8
$ conda create --prefix ~/llm python=3.11 pip --y
```
**Note:** "~" in linux means home directory, it is similar to "/users/yourusername"
Once done, activate conda env and install dependency:

```
$ conda activate ~/llm
$ pip install ipykernel transformers==4.46.1 sentencepiece sacremose pandas diffusers accelerate safetensors
$ pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
```


## 2. Jump host to SuperPOD compute node
### 2.1 Install Remote - SSH
- Click File\New Window (Ctrl+Shift+N)
- Install Remote - SSH

    - Go to the Extensions panel (Ctrl+Shift+X)
    - Search for "Remote - SSH" (by Microsoft)
    - Click Install

### 2.2 Connect to a Remote Host:

- Press Ctrl+Shift+P/or click **Connect via SSH** to open the Command Palette
- Type **"Remote-SSH: Connect to Host..."** and select it
- Click **"Configure SSH Host ..."** to open config file
- Enter the following script to the config file:

```
Host superpod
    HostName superpod.smu.edu
    User username

Host GPU
    HostName bcm-dgxa100-0001
    User username
    ProxyJump superpod
```

- Note: enter your own username. The HostName needs to change to the correct GPU nodes that you requested in previous step.
- Click on Connect to GPU
- (Select Linux for Select platform of Remote Host GPU)
- Enter Password and Duo for login node and Password again for Compute node.
- You can also select to browse the files on SuperPOD and enter Password/Duo to authenticate it.
- Once authenticated, you can work with any Jupyter Notebook from the SuperPOD directory
