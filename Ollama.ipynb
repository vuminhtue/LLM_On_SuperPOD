# Ollama
## Load Ollama

```
$ module load testing/ollama
$ ollama --version
```

## Serve Ollama onto terminal 
- Using **&** to make Ollama run in background
```
$ ollama serve &
```

## List downloaded models:

```
$ ollama list
```

## Download model
- pull model name:version
```
$ ollama pull gemma3:1b
```

## Run model

```
$ ollama run gemma3:1b
```

## Stop Ollama

```
$ killall ollama
```

## Running Ollama in Jupyter Notebook
### Find the Ollama port
- It is important to find the right port that ollama serve when running it on SuperPOD
- The reason behind is that a node in SuperPOD has 8 GPUs and it is possible that you are sharing the same node with other 7 people. If each person serve 1 ollama on their GPU, there will be a glitch
- Thus when using ollama module preloaded on SuperPOD, one need to find the right port that serve ollama:

````python
import os
from langchain_community.chat_models import ChatOllama

# get home and job id
home_dir =  os.getenv('HOME')
job_id = os.getenv('SLURM_JOB_ID')

# get ollama directory or default to home
ollama_dir = os.getenv('OLLAMA_BASE_DIR', home_dir)

try:    
    with open(f"{ollama_dir}/ollama/host_{job_id}.txt") as f:
        HOST = f.read().strip()
    with open(f"{ollama_dir}/ollama/port_{job_id}.txt") as f:
        PORT = f.read().strip()
    ollama_url = f"http://{HOST}:{PORT}"
except Exception as e:
    print("[⚠️] Could not read host/port, you manually set the `ollama_url` below.")

print("The port that serve ollama is: ", ollama_url)
````

````python
chat_model = ChatOllama(
            base_url=ollama_url,
            model="gemma3:1b",
            temperature=0,
        )

response = chat_model.invoke("Where is SMU?")
print(response.content)
````

````python
response = chat_model.invoke("Write a poem about SMU in Dallas?")
print(response.content)
````
